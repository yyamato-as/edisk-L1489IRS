## Connection to ERDA via SFTP
for more details see https://ext.erda.dk/wsgi-bin/setup.py?topic=sftp 
1. Confirm that something like the following lines in your local ~/.ssh/config to avoid typing the full login details every time:

```
Host io.erda.dk ucph-erda
Hostname io.erda.dk
VerifyHostKeyDNS yes
User yamato@astron.s.u-tokyo.ac.jp
Port 22
# Uncomment next line to use your private key in ~/.ssh/id_rsa
# IdentityFile ~/.ssh/id_rsa
```

2. Connect to ERDA with the following command in a terminal

```
sftp -B 258048 io.erda.dk
```

You may asked to enter the passphrase for your ssh key.

## Download data from ERDA server
After connection to the ERDA server, you can browse as in usual terminal. To download a certain file or folder, execute the following command:

```
sftp> get -r [filename] [path of local directory where you want to download the file]
```

You may do the command above on a screen environment, since usually the ALMA data is huge and take time to download.
