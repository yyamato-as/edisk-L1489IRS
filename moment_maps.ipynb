```python
from qdisk.product import calculate_moment
from qdisk.plot import Map
from qdisk.classes import FitsImage
import analysis_utils as au
from eDisk_source_dict import source_dict
from eDiskplot import cpal as freeze
from astropy.visualization import AsinhStretch, SinhStretch, ImageNormalize, LinearStretch
from matplotlib import ticker
import numpy as np
from matplotlib.cm import ScalarMappable
%matplotlib widget
%load_ext autoreload
%autoreload 2

source = "L1489IRS"
config = "SBLB"
line = "SO"
robust = 1.0
dv = 0.2
vel_extent = source_dict[source]["emission_extent"][line]
center_coord = source_dict[source]["radec"]
vsys = source_dict[source]["vsys"]
imagename = au.customimagepath + au.get_image_basename(source=source, baseline=config, line=line, robust=robust, dv=dv)
```

```python
# mom0
mom0name = au.VADPpath + au.get_image_basename(source=source, baseline=config, line=line, robust=robust, dv=dv)
```

```python
threshold = 2
calculate_moment(
    imagename=imagename,
    moments=[0],
    threshold=threshold,
    vel_extent=vel_extent,
    savefilename=mom0name,
)
```

```python
# plot
rmax = 3
mom0map = Map(mom0name.replace(".fits", "_M0.fits"), center_coord=center_coord, xlim=(-rmax, rmax), ylim=(-rmax, rmax))
mom0map.plot_colormap(cmap=freeze, vmin=0.0, method="contourf", levels=16, extend="min")#, stretch=AsinhStretch())
mom0map.add_beam()
mom0map.add_colorbar(label="mJy beam$^{-1}$ km s$^{-1}$")
mom0map.colorbar.ax.set_ylim(0.0, None)
# mom0map.colorbar.set(extend="min")
# norm = ImageNormalize(
#             mom0map.data, vmin=0.0, vmax=None, interval=None, stretch=LinearStretch()
#         )
# mom0map.ax.get_figure().colorbar(ScalarMappable(norm=norm, cmap=freeze))
# mom0map.colormap.norm.vmin
# np.nanmin(mom0map.data) < mom0map.colormap.norm.vmin
# mom0map._set_colorbar_extend()
# mom0map.add_colorbar()
# mom0map.colorbar.extend
# np.nanmin(mom0map.data)
# mom0map.colormap.get_clim()
```

```python
# velocity map
v0name = au.VADPpath + au.get_image_basename(source=source, baseline=config, line=line, robust=robust, dv=dv)
```

```python
calculate_moment(
    imagename=imagename,
    moments=["q"],
    vel_extent=vel_extent,
    savefilename=v0name,
)
```

```python
# plot
rmax = 3
vrange = 4
# mask the region where the spectrum peak is below 4sigma
cube = FitsImage(imagename, xlim=(-rmax, rmax), ylim=(-rmax, rmax))
cube.shift_phasecenter_toward(center_coord)
print("Generating threshold mask...")
cube.estimate_rms(edgenchan=3)
cube.get_threshold_mask(threshold=3.5)
mask = np.sum(cube.threshold_mask, axis=0) > 0
print("Done.")

# plot
v0map = Map(v0name.replace(".fits", "_v0.fits"), center_coord=center_coord, xlim=(-rmax, rmax), ylim=(-rmax, rmax), data_scaling_factor=1e-3)
v0map.mask(vmin=vsys-2, vmax=vsys+2, user_mask=mask, combine="or")
v0map.plot_colormap(cmap="RdBu_r", method="pcolorfast", vmin=vsys-vrange, vmax=vsys+vrange)
v0map.add_beam()
v0map.add_colorbar()
# v0map.colorbar.ax.set_ylim(vsys-vrange, vsys+vrange)
```
